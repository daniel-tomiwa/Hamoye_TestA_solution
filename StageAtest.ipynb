Importing the dataset and reading it into a pandas Dataframe

```python
Food_n_agric = pd.read_csv("FoodBalanceSheets_E_Africa_NOFLAG.csv", encoding="iso-8859-1")
```

```python
Food_n_agric.head()
```

11. What is the total sum of Animal Fat produced in 2014 and 2017 respectively?

```python
Food_n_agric.groupby("Item")["Y2014", "Y2017"].sum().loc["Animal fats"]
```

12. What is the mean and standard deviation across the whole dataset for the year 2015 to 3 decimal places?

```python
Food_n_agric.describe()["Y2015"][["mean", "std"]]
```

13. What is the total number and percentage of missing data in 2016 to 2 decimal places?

```python
missing_values_2016 = Food_n_agric.isnull().sum()["Y2016"]
```

```python
Total_values_2016 = Food_n_agric["Y2016"].shape[0]
Total_values_2016
```

```python
percentage_null_2016 = missing_values_2016 / Total_values_2016 * 100
print(percentage_null_2016)
```

14. Which year had the highest correlation with ‘Element Code’?

```python
abs(Food_n_agric.corr()["Element Code"]).sort_values(ascending=True)
```

15. What year has the highest sum of Import Quantity?

```python
Food_n_agric.groupby("Element").sum()[["Y2014", "Y2015", "Y2016", "Y2017", "Y2018"]].loc["Import Quantity"]
```

16. What is the total number of the sum of Production in 2014?

```python
Food_n_agric.groupby("Element").sum()["Y2014"]
```

17. Which of these elements had the highest sum in 2018?

```python
Food_n_agric[["Element", "Y2018"]].groupby("Element").sum()
```

18. Which of these elements had the 3rd lowest sum in 2018?

```python
Food_n_agric[["Element", "Y2018"]].groupby("Element").sum().sort_values(by="Y2018", ascending=True)
```

19. What is the total Import Quantity in Algeria in 2018?

```python
Food_n_agric.groupby(['Element', 'Area'])['Y2018'].aggregate('sum').unstack()
```

20. What is the total number of unique countries in the dataset?

```python
no_of_countries = len(Food_n_agric["Area"].unique())
print(no_of_countries)
```
